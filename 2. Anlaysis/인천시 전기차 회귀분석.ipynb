```python
df['총/기간'] = [0]*len(df)
df['총/빈도'] = [0]*len(df)
df['총/일수'] = [0]*len(df)

# 서비스 기간이 값이 0인 행의 index
a = set(df[ df['서비스기간'] == 0 ].index)

# 서비스 빈도가 값이 0인 행의 index
b = set(df[ df['서비스빈도'] == 0 ].index)

# 서비스 일 수가 값이 0인 행의 index
c = set(df[ df['서비스일수'] == 0 ].index) 

# 위에서 구한 0인 값들 제외하고 기준에 따른 값들 구하기

df.loc[ list(set(df['서비스기간'].index) - a), '총/기간'] = (
    df.loc[ list(set(df['서비스기간'].index) - a), '총_충전량'] / df.loc[ list(set(df['서비스기간'].index) - a), '서비스기간']
)
    
df.loc[ list(set(df['서비스빈도'].index) - b), '총/빈도'] = (
    df.loc[ list(set(df['서비스빈도'].index) - b), '총_충전량'] / df.loc[ list(set(df['서비스빈도'].index) - b), '서비스빈도'] 
)
    
df.loc[ list(set(df['서비스일수'].index) - c), '총/일수'] = (
    df.loc[ list(set(df['서비스일수'].index) - c), '총_충전량'] / df.loc[ list(set(df['서비스일수'].index) - c), '서비스일수'] 
)

df
```
