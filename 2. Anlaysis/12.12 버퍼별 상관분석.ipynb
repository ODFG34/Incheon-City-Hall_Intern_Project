### Column별 MinMax 정규화

```python
df_g2 = df_g.copy()
for i in df_g2.columns[1:] :
    df_g2[i] = df_g2[i] / (df_g2[i].max() - df_g2[i].min())
df_g2.head()
```

```python
df_temp = df_g2.loc[:, ['gid', '충전량_kWh', '생산가능인구', '교통량', '변압기', '도서관', '민행기_SN', '음식점',
                       '공원', '경기장', '대형마트', '학교', '주차면수', '전기차']]
df_temp.sort_values(['생산가능인구'], ascending = False)
```
