## 충전일수 충전량 도서관 변압기 전기차 경제인구수 주차면수 공원 
## 음식점 학교 민행기_ SN 경기장 대형마트 교통량 건물1 건물2 건물3

```python
#'충전일수', '충전량_kWh', '도서관', '변압기', '전기차', '경제인구수', '주차면수', '공원', 
#'음식점', '학교', '민행기_ SN', '경기장' ,'대형마트', '교통량' '건물1', '건물2' , '건물3'
```

```python
col_li = ['충전일수', '충전량_kWh', '도서관', '변압기', '전기차', '경제인구수','주차면수', '공원', '음식점', '학교', 
          '민행기_SN', '경기장' ,'대형마트', '교통량']
col_li
```

```python
df2 = df.loc[:, col_li]
df2.head(1)
```

```python
df2['일당충전량'] = df2['충전량_kWh'] / df2['충전일수']
df2.head(1)
```

```python
df2.corr()
```

## 총충전량과 변수들의 상관계수

```python
weight1 = pd.DataFrame(df2.corr().iloc[1, 2:-1])
weight1.rename(columns = {
    '충전량_kWh' : '충전량과 상관계수'
}, inplace = True)
weight1
```

## 일당충전량과 변수들의 상관계수

```python
weight2 = pd.DataFrame(df2.corr().iloc[-1, 2:-1])
weight2.rename(columns = {
    '일당충전량' : '일당충전량과 상관계수'
}, inplace = True)
weight2
```

# 격자 데이터 가져오기

```python
col_li.insert(0, 'gid')
```

```python
gdf = pd.read_csv("인천시_격자_g인도변전충주학민SN공음경마교.csv", encoding = "CP949")
print(gdf.shape)
gdf.head(1)
```

```python
gdf = gdf.loc[:, col_li]
```

```python
gdf.head()
```

```python
gdf_nor = gdf.copy()

for i in gdf_nor.columns[1:] :
    gdf_nor[i] = gdf_nor[i] / (gdf_nor[i].max() - gdf_nor[i].min())
gdf_nor.sort_values(['경제인구수'], ascending = False)
```

```python
Score1 = pd.DataFrame(np.array(gdf_nor.iloc[:,3:]).dot(np.array(weight1)))
Score1
```

```python
Score2 = pd.DataFrame(np.array(gdf_nor.iloc[:,3:]).dot(np.array(weight2)))
Score2
```

```python
gdf['Score1'] = Score1
gdf['Score2'] = Score2
```

```python
gdf.sort_values(['Score1', 'Score2'], ascending = False)
```
