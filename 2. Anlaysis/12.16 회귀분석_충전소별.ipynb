```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("1500M버퍼_충전소별 정보.csv", encoding = "CP949")
df.head(3)
```

```python
x_li = list(df.columns[4:-1])
x_li
```

```python
df2 = df.loc[ :, x_li]
```

```python
import statsmodels.api as sm

s= '일당충전량 ~ '
for i in x_li :
    if i != '경기장':
        s += '+' +  i
    else :
        s += i

reg = sm.OLS.from_formula(s, data = df).fit()
reg.summary()
```
