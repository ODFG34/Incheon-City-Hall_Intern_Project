```python
import pandas as pd
import numpy as np
```

### 충전소 버퍼 2km 내의 각 변수들의 개수

```python
df_cor = pd.read_csv("인천시_2km_관공서_변압기_전기차_경제인구.csv", encoding = "CP949")
df_cor.head(1)
```

```python
df_cor2 = df_cor.loc[:, ['충전소','주소', '관공서', '변압기', '전기차', '경제인구', '충전량(kWh', '충전일수']]
df_cor2.head(1)
```

```python
df_cor2['일당충전량'] = df_cor2['충전량(kWh'] / df_cor2['충전일수']
df_cor3 = df_cor2.iloc[:, [0,1,2,3,4,5,6,8,7]]
df_cor3.head(2)
```

### 격자 내의 각 변수들의 개수

```python
df = pd.read_csv("19.12.03_격자의 변수별 개수.csv", encoding = "CP949")
df.head(1)
```

```python
df2 = df.loc[:, ['관공서', '변압기', '전기차', '경제인구수','총충전량']]
```

# 상관관계 도출

```python
# 주차면 변수 없음
value = df_cor3.iloc[:,2:-2].corr()
value = pd.DataFrame(value.iloc[-1, :])
print(value.shape, type(value))
value
```

# 각 변수들의 값을 이용하여 Score 값 추출

## 변수 :  경제인구    관공서(민원행정기관 + 도서관)    변압기    전기차    주차면

df => (n X 5) * (5 X 1) => n X 1 => n개의 격자들의 Score 행

### Column별 MinMax 정규화

```python
df3 = df2.copy()
for i in df2.columns :
    df3[i] = df2[i] / (df2[i].max() - df2[i].min())
df3.head()
```

```python
df3.sort_values(['경제인구수'], ascending = False)
```

```python
SS = pd.DataFrame( np.array(df3).dot(np.array(value)))
SS
```

```python
df3['Score'] = SS
```

```python
df3.head(1)
```

```python
df3.sort_values(['Score'], ascending = False)
```
