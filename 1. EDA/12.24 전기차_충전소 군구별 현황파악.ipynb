```python
import pandas as pd
import numpy as np
```

```python
ch = pd.read_csv("충전소명_주소.csv", encoding = "CP949")
```

```python
ch
```

```python
ch.columns
```

```python
ch2 = ch.iloc[:, [1,2]]
```

```python
ch2
```

```python
ch2['군구'] = list(map(lambda x: x.split(' ')[1], ch2['주소']))
ch2
```

```python
ch2[ch2['충전소'] == '롯데마트 영종도점']
```

```python
ch2 = ch2.drop_duplicates(['충전소'], keep = 'first')
```

```python
ch2
```

```python
s1 = pd.DataFrame(ch2.groupby(['군구']).size(), columns = ['수'])
s1
```

```python
s1['군구'] = list(s1.index)
s1.index = range(len(s1))
s1 = s1.iloc[:, [1,0]]
s1
```

```python
s1.to_csv("12.24 군구별 충전소 수.csv", encoding = "CP949", index = False)
```
