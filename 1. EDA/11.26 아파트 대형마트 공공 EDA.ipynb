```python
mart_li = ['홈플러스', '마트' ]
temp_df = []

for i in range(len(hj2)) :
    if( (hj2.loc[i, '충전소'].find(mart_li[0]) != -1) | (hj2.loc[i, '충전소'].find(mart_li[1]) != -1) ) :
        temp_df.append( hj2.iloc[i,:])
        #temp_df.append(  (pd.DataFrame(hj2.iloc[i,:], index= hj2.columns).T), ignore_index=True)

for i in range(len(hk)) :
    if( (hk.loc[i, '충전소'].find(mart_li[0]) != -1) | (hk.loc[i, '충전소'].find(mart_li[1]) != -1) ) :
        s = hk.iloc[i,:].copy()
        s['시작일시'] = str(s['시작일시'])
        s['시작일시'] = s['시작일시'][:4] + '-' + s['시작일시'][4:6] + '-' +s['시작일시'][6:8] + ' '+ s['시작일시'][8:10]+ ':' + s['시작일시'][10:12]+':' + s['시작일시'][12:14]
        temp_df.append(s)        

temp_df = pd.DataFrame(temp_df, columns = hj2.columns)
temp_df.index = range(len(temp_df))
print(len(temp_df), temp_df['충전소'].unique() )
temp_df.head()
```

```python
temp_df.head(1)
```

```python
temp_df['시작일시']
```

```python
temp_df['시작일시'][1120].split(' ')[1].split(':')[0]
```

```python
temp_df['시간대'] = list(map(lambda x : x.split(' ')[1].split(':')[0], temp_df['시작일시']))

temp_df2 = temp_df.groupby(['시간대']).sum()
temp_df2['시간대'] = temp_df2.index
temp_df2 = temp_df2.iloc[:,[2,1]]
temp_df2
```

```python
len(temp_df['충전소'].unique())
```

```python
n_li = list(temp_df.groupby(['시간대']).size())
n_li
```

```python
temp_df2['평균충전량'] = temp_df2['충전량(kWh)'] / len(temp_df['충전소'].unique())
temp_df2
```

```python
qqq2 = qqq.groupby(['시간대']).agg('std').fillna(0)
qqq2.rename( columns = {
    '충전량(kWh)' : '충전량(kWh) 편차'
}, inplace = True)
```

```python
qqq2
```

```python
pd.Series(list(qqq2.iloc[:, 0])) / n_li
```

```python
n_li
```

```python
temp_df3 = temp_df2.copy()
temp_df3.index = range(len(temp_df3))
temp_df3['편차'] = list(qqq2.iloc[:, 0])
temp_df3['MSE'] = list(temp_df3['편차'] / n_li)
temp_df3['시간대'] = list(map(lambda x : str(x) + ":00", temp_df3['시간대']))
temp_df3
```
