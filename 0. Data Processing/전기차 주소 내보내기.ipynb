```python
hanjeon = pd.read_excel("급속충전이용내역_한전_18.01-19.10.xlsx")
hj = hanjeon.copy()
hj.head(3)
```

```python
hwangyeong = pd.read_excel("급속충전이용내역_환경부_18.01-19.10.xlsx")
hg = hwangyeong.copy()
hg.head(3)
```

```python
hg.rename( columns = {
    '충전소명' : '충전소'
},inplace = True)
```
