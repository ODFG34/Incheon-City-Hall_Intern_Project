```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("인천시_gis건물_일반집합_주차면50이상_용도코드주차면.csv", encoding = "CP949")
df.head(1)
```

```python
ori_li = df.columns
aft_li = ['식별번호', '용도코드', '용도명', '주차면수']
```

```python
df.columns = aft_li
```

```python
df.head(1)
```

```python
df['용도명'].unique()
```

```python
df['용도코드'].unique()
```

```python
df['용도코드및명'] = list(map(lambda x,y : x + ' ' + y, df['용도코드'], df['용도명']))
df.head(1)
```

```python
len(df['용도코드및명'].unique())
```

```python
temp = pd.DataFrame( [df['용도코드및명'].unique()], index = ['용도코드및명']).T
temp
```

```python
temp['용도코드'] = list(map(lambda x: x.split(' ')[0], temp['용도코드및명']))
temp['용도명'] = list(map(lambda x: x.split(' ')[1], temp['용도코드및명']))
temp
```

```python
temp2 = temp.iloc[:, 1:]
temp2
```

```python
temp3 = temp2.sort_values(['용도코드'])
temp3.index = range(len(temp3))
temp3
```

```python
temp3['건물가점'] = [1,3,2,2,2,
                 1,1,2,2,2,
                 1,2,3,0,2,
                0,0,0,2,0,
                0,0,1,0,2,
                3,2]
temp3
```

```python
Tot = pd.merge( df, temp3, on = '용도코드', how = 'left')
Tot
```

```python
Tot2 = Tot.loc[: ,['식별번호', '용도코드', '용도명_x', '주차면수', '건물가점']]
Tot2
```

```python
ori_li = list(ori_li)
ori_li.append('BuildWeight')
Tot2.columns = ori_li
Tot2
```
